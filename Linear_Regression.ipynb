```python
import numpy as np
import matplotlib.pyplot as plt

data_set = np.loadtxt("C:\ex1data1.txt", delimiter = ",")

x = data_set[:, 0] # x에 배열 첫 번째 열 저장
y = data_set[:, 1] # y에 배열 두 번째 열 저장

theta_0 = np.random.rand()
theta_1 = np.random.rand()

def h(x): #가설함수
    return theta_0 + theta_1 * x 

def J(x, y): #손실함수(제곱 오차의 합)
    return (1/2) * np.sum((y - h(x)) **2) # 주어진 데이터 결과값과 가설함수의 오차를 제곱. 1/2 곱은 미분 값 가독성을 위해

learning_rate = 0.0001 # 학습률. 가장 

# 변화율 -> 가설함수의 기울기
dev = 1 # 최초 변화율 임의 설정 1. 0에 근접하게끔 줄인다.

count = 1 #반복하는 회수. 손실함수 미분해서 theta_0과 theta_1의 갱신 반복하는 횟수

error = J(x, y)                                                 # 실제 값(data_set)을 넣어서 가설함수와 손실함수의 오차를 error에 저장
                                                                # (처음 오차)
while dev > 0.001:                                              # 변화율이 0에 근접할 때까지 while문을 반복.
    temp_0 = theta_0 - learning_rate*np.sum(h(x) - y)          # 손실함수를 theta_0에 대해 미분하고 실제값을 넣어서 갱신값을 temp_0에 저장
    temp_1 = theta_1 - learning_rate*np.sum((h(x) - y) * x)    
    theta_0 = temp_0                                           # 갱신되어 저장한 값을 다시 변화율을 0에 근접하게 하기 위해 theta_0에 저장 
    theta_1 = temp_1
    error_next = J(x, y)                                       # 갱신되어 저장된 값으로 만들어진 오차(나중 오차)
    dev = error - error_next                                   # 변화율 = 처음오차 - 나중 오차
    error = error_next                                         # 갱신 반복을 위해 나중 오차를 처음 오차에 저장
    
plt.scatter(x, y) # x, y에 저장된 데이터들을 테이블 위에 뿌리기

plt.xlabel("x(Feature value)")
plt.ylabel("y(Target value)")
plt.title("Data set(Target value & Linear regression)")
plt.plot(x, h(x), color = 'red') # 데이터들을 가장 잘 표현해 줄 선형 회귀선 그리기.

# 평균 제곱 오차(MSE) 계산 및 출력
y_pred = h(x) # 예측값 계산
mse = np.mean((y - y_pred) ** 2) # 제곱 오차들에 .mean 함수 이용하여 평균 제곱 오차 계산.
print("평균 제곱 오차(MSE):", mse) # 평균 제곱 오차 출력

```

```python
# 선형 회귀 결과와 실제 값을 비교하는 표 출력
plt.scatter(x, y, label='real data')
plt.plot(x, y_pred, color='red', label='Linear regression line')
plt.xlabel("x(Feature value)")
plt.ylabel("y(Target value)")
plt.title("Data set(Target value & Linear regression)")
plt.legend()
plt.show()
```
